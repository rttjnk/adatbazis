```sql
-- kinek nincs telefonszm megadva

SELECT first_name + ' ' + last_name as 'Name'
FROM customers
WHERE phone is NULL
```

```sql
-- megadott-e telefonszmot

SELECT van_e,
count(*) as 'darabszm'
FROM 
    (select 
    case when phone is null then 'Nincs megadva'
        else 'Meg van adva'
        end as van_e
    FROM customers) rows
GROUP BY van_e
```

```sql
-- telefonszmok csoportostsa

SELECT SUBSTRING(phone,2,3) as 'Krzetazonost', COUNT(customer_id) as 'Darabszm'
from customers
WHERE SUBSTRING(phone,2,3)  is not NULL
group by SUBSTRING(phone,2,3) 
order BY SUBSTRING(phone,2,3) 
```

```sql
-- vrosok llamonknt

SELECT state as 'llam', count(city) as 'Vros/llam'
from customers
group BY state
```

```sql
-- leggyakoribb vrosok

select top 10
    city as 'Vros',
    count(*) as 'Darab/vros'
From customers
WHERE city is not NULL
GROUP by city
order BY 'Darab/vros' DESC
```

```sql
-- Egy szbl ll-e a vros neve?

SELECT Egyszavas,
count(*) as 'Darab'
FROM 
    (select 
    case when charindex(' ',city) = 0 then 'Igen'
        else 'Nem '
        end as Egyszavas
    FROM customers) rows
GROUP BY Egyszavas
```

```sql
-- leghosszabb, legrvidebb vrosnv, tlagos vrosnv-hossz

select max(LEN(city)) as 'Leghosszabb vrosnv karaktereinek szma',
        MIN(LEN(city)) as 'Legrvidebb vrosnv karaktereinek szma',
        avg(len(city)) as 'tlagos vrosnv-hossz karakterben'
from customers
```

```sql
-- leghosszabb vros neve

select city as 'Leghosszabb nev vros(ok)'
From customers
GROUP by city
HAVING LEN(city) = 22

```

```sql
--legrvidebb vrosnv

select city as 'Legrvidebb nev vros(ok)'
From customers
GROUP by city
HAVING LEN(city) = 4
```
