```sql
-- kinek nincs telefonszám megadva

SELECT first_name + ' ' + last_name as 'Name'
FROM customers
WHERE phone is NULL
```

```sql
-- megadott-e telefonszámot

SELECT van_e,
count(*) as 'darabszám'
FROM 
    (select 
    case when phone is null then 'Nincs megadva'
        else 'Meg van adva'
        end as van_e
    FROM customers) rows
GROUP BY van_e
```

```sql
-- telefonszámok csoportosítása

SELECT SUBSTRING(phone,2,3) as 'Körzetazonosító', COUNT(customer_id) as 'Darabszám'
from customers
WHERE SUBSTRING(phone,2,3)  is not NULL
group by SUBSTRING(phone,2,3) 
order BY SUBSTRING(phone,2,3) 
```

```sql
-- városok államonként

SELECT state as 'Állam', count(city) as 'Város/állam'
from customers
group BY state
```

```sql
-- leggyakoribb városok

select top 10
    city as 'Város',
    count(*) as 'Darab/város'
From customers
WHERE city is not NULL
GROUP by city
order BY 'Darab/város' DESC
```

```sql
-- Egy szóból áll-e a város neve?

SELECT Egyszavas,
count(*) as 'Darab'
FROM 
    (select 
    case when charindex(' ',city) = 0 then 'Igen'
        else 'Nem '
        end as Egyszavas
    FROM customers) rows
GROUP BY Egyszavas
```

```sql
-- leghosszabb, legrövidebb városnév, átlagos városnév-hossz

select max(LEN(city)) as 'Leghosszabb városnév karaktereinek száma',
        MIN(LEN(city)) as 'Legrövidebb városnév karaktereinek száma',
        avg(len(city)) as 'Átlagos városnév-hossz karakterben'
from customers
```

```sql
-- leghosszabb város neve

select city as 'Leghosszabb nevű város(ok)'
From customers
GROUP by city
HAVING LEN(city) = 22

```

```sql
--legrövidebb városnév

select city as 'Legrövidebb nevű város(ok)'
From customers
GROUP by city
HAVING LEN(city) = 4
```
